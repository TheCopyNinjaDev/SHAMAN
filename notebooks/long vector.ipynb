```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn')
from tqdm import tqdm
```

```python
df = pd.read_csv(f'../data/U/2020-10-01 - 2021-10-02.csv', index_col=0, parse_dates=['time'], dayfirst=True)
df.head()
```

```python
plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(df['c'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
test_size = 74
week_size = test_size * 6
data = df.filter('c')
dataset = data.values
train_df = dataset[:-test_size]
test_df = dataset[-test_size:]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(train_df)
scaled_data
```

```python
x_train = []
y_train = []

for i in range(week_size, len(scaled_data), test_size):
    x_train.append(scaled_data[i - week_size:i])
    y_train.append(scaled_data[i:i+test_size])

x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model=Sequential()
model.add(LSTM(444,return_sequences=True,input_shape=(week_size,1)))
model.add(LSTM(74,return_sequences=True))
model.add(LSTM(37))
model.add(Dropout(0.20))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
%%time

model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
x_test = x_train[len(x_train) - 1:].copy()
predictions = []

for i in range(len(test_df)):
    predictions.append(model.predict(x_test).tolist())
    x_test = np.append(x_test[0][1:], predictions[i])
    x_test = x_test.reshape(1, week_size, 1)
```

```python
predictions = np.array(predictions)
predictions = predictions.reshape(74, 1)
predictions = scaler.inverse_transform(predictions)
```

```python
train = data[:-test_size]
valid = data[-test_size:]
valid['Predicted'] = predictions
valid
```

```python
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['c'])
plt.plot(valid[['c', 'Predicted']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
